```
#Genera regresión lineal a partir de dos vectores(x,y) de igual longitud
import numpy as np                                              #Librerias de manejo números y de arrays
import pandas as pd                                             #Libreria para manejo de archivos
import matplotlib.pyplot as plt                                 #Librería para crear y mostrar gráficos
Datos=pd.read_csv("InformacionSalarios.csv")
x=np.array(Datos.sort_values('annosExperienca').annosExperienca[1:30].tolist())     #Sort-ordena, Edad[n:m]-Corta desde n hasta m, tolist()-lo convvierte a array generico  
y=np.array(Datos.sort_values('salario').salario[1:30].tolist()) 
n=len(x)                                                        #Longitud del vector que se va a trabajar, sirve para calcular promedios
sumax= sum(x)                                                   #Suma todas las edades
sumx=sum(x)
sumy=sum(y)                                                     #Suma todos los salarios                                                         
sumx2=0                                                         #crea los acomuladores 
sumy2=0
sumyx=0
promx=sumx/n                                                    #Calcula promedios
promy=sumy/n
for i in ([0,n-1]):                                             #Repetirá los proximos tres renglones 
   sumx2=sumx2+(x[i]*x[i])                                      #Calcula el cuadrado de Edad[i] y lo suma al acomulador  
   sumy2=sumy2+(y[i]*y[i])                
   sumyx=sumyx+(y[i]*x[i])                                      #Calcula Edad[i]* Salario[i]
m=(sumx*sumy-n*sumyx)/(sumx**2-n*sumx2)                         #Calcula la pendiente 
b=promy-m*promx

print ("La función lineal de la regresión es: Y = "+str(m)+"x + " + str(b))
                                            #Calcula b, el desplazamiento de la linea
plt.plot(x, m*x+b, label="Ajuste")           #Creamos la gráfica                             #Crea la linea y= mx+b , y le asigna el nombre ajuste
plt.plot(x,y,marker='o', label="Datos")                    #Crea la grafica de los puntos                    #Muestra los puntos (Edad, Salario)
plt.legend(loc=4)                                               #Ubica la etiqueta en la cuarta esquina
plt.show()                                                      #Muestra el gráfico 
```
